Let's talk ``pickle``. From the Python documentation and the Internet, pickle one of the previously used files. Once it's nice and briny, unpickle it and show its format. Describe what ``pickle`` is and what it does. (OPTIONAL)

```python
import pickle
## code adapted from class notes from summer python class at ucb extension
data4 = pd.read_csv("data4.txt", names=column_names, sep="$") # read in file

try:
    pickled_d4 = open('data4.dump', 'wb')
    
except IOError: 
     
    sys.stderr.write('Error opening employees.dump for write') 
    sys.exit(1)

pkl = pickle.Pickler(pickled_d4) 

pkl.dump(data4) 
pickled_d4.close() 
```

```python
# When it is needed in another file, add code such that the following aspects are included in order:

import pickle

pickled_d4 = open('data4.dump', 'rb')

data4 = pickle.load(pickled_d4)

# From here, you can use indexes to extract specific information from the data.

display(data4.head()) # going back to the format of the above problems to demonstrate unpickling
```

``pickle`` is a bit of an odd bird within Python. Not all object types can be pickled (for example, generator and lambda functions cannot be), and it is not necessarily usable outside Python or cross-compatible among different versions of the language. It works by converting the object in question into bytes (binary form) for more convenient storage using the ``Pickler`` method, and may be used across files.

Light as a ``feather``, but with less complex air resistence, the feather object type from Python is another type of serialization function (like the ``pickle`` mentioned above).
